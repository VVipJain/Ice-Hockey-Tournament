```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

finaldf = pd.DataFrame()

for i in range(1, 7):
    url = f'https://www.scrapethissite.com/pages/forms/?page_num={i}&per_page=100'
    webpage = requests.get(url, headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    # Extracting table headers
    hockey_titles = soup.find_all('th')
    hockey_table_titles = [title.text.strip() for title in hockey_titles]
    
    # Creating a DataFrame with the headers
    df = pd.DataFrame(columns=hockey_table_titles)
    
    # Extracting row data
    row_data = soup.find_all('tr')
    for row in row_data[1:]:  # Assuming the first row is the header
        column_data = row.find_all('td')
        individual_column_data = [data.text.strip() for data in column_data]
        df.loc[len(df)] = individual_column_data  # Append row to the DataFrame
    
    # Append the extracted DataFrame to the final DataFrame using concat
    finaldf = pd.concat([finaldf, df], ignore_index=True)

# Display final dataframe
print(finaldf)
```

```python
finaldf.to_csv('hockey-tournament-data.csv',index=False)
```
